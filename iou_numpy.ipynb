交并比 IoU

```python
import numpy as np
```

```python
# Box coordinates: [x1, y1, x2, y2]
box1 = np.array([50, 50, 150, 150])  
box2 = np.array([100, 100, 200, 200])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 黑色背景, 矩形
image = np.zeros((300, 300, 3), dtype=np.uint8)

# box1, 蓝色, 左上角, (x1,y1)=(50, 50)
# 蓝色框的右下角 (x2, y2) = (150, 150)
image[box1[1]:box1[3], box1[0]:box1[2], 2] = 255

# box2, 红色框, 左上角, (x1,y1)=(100, 100)
# 红色框的右下角 (x2, y2) = (200, 200)
image[box2[1]:box2[3], box2[0]:box2[2], 0] = 255

plt.imshow(image)
plt.show()
```

```python
def get_iou(box1, box2):
    # 交集 intersection 区域
    x1 = max(box1[0], box2[0]) # 交集左上角横坐标 x
    y1 = max(box1[1], box2[1]) # 交集左上角纵坐标 y
    x2 = min(box1[2], box2[2]) # 交集右下角横坐标 x
    y2 = min(box1[3], box2[3]) # 交集右下角纵坐标 y

    # 交集区域面积
    intersection_area = max(0, x2 - x1) * max(0, y2 - y1)
    box1_area = (box1[2] - box1[0]) * (box1[3] - box1[1]) # 蓝色框面积
    box2_area = (box2[2] - box2[0]) * (box2[3] - box2[1]) # 红色框面积

    iou = intersection_area / (box1_area + box2_area - intersection_area + 1e-6)
    return iou
```

```python
get_iou(box1, box2)
```
